```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
from requests import get
from bs4 import BeautifulSoup
```

```
url = "https://spotifycharts.com/regional/global/weekly/latest"
html = get(url)
```

```
soup = BeautifulSoup(html.text, 'html.parser')
type(soup)
```

```
title = soup.title
print(title)
```

```
text = soup.get_text()
print(soup.text)
```

```
#this is the container that stores all the data that I want e.g name of the artist, song and the number of streams
main_content=soup.find_all('div', class_='container')
print(type(main_content))
```

```
#This tells me the name of the artist
soup.find_all('span')
```

```
#This tells me the name of the song
soup.find_all('strong')
```

```
chart_table=soup.find_all('table', class_='chart-table')
```

```
rows=soup.find_all('tr')
print(rows[:11])
```

```
for row in rows:
    row_td = row.find_all('td')
    str_cells = str(row_td)
    cleantext = BeautifulSoup(str_cells, "lxml").get_text()
    print(cleantext)
print(row_td)
```

```
import re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells).replace('\\n','')
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
    print(clean2)
type(clean2)
```

```
df = pd.DataFrame(list_rows)
df.head(11)
```

```
df1 = df[0].str.split(',', expand=True)
df1.head(11)
```
